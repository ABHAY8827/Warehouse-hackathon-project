### Model building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_percentage_error
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
for i in range(0,1000):
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state = i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train) 
    pred_test = lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1) == round(r2_score(y_test,pred_test)*100,1):
        print("At random state",i,"model perform very well")
        print("At random state",i,",")
        mape = mean_absolute_percentage_error(y_test,pred_test)
        score = max(0, 100*(1-mape))
        print('Score :', score)
        mse = mean_squared_error(y_test,pred_test)
        print('RMSE :', np.sqrt(mse))
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
mape = mean_absolute_percentage_error(y_test,pred)
score = max(0, 100*(1-mape))
print('Score :', score)
print("mean_absolute_error",mean_absolute_error(y_test,pred))
mse = mean_squared_error(y_test,pred)
print('RMSE :', np.sqrt(mse))
```

## Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
pred = rf.predict(x_test)
mape = mean_absolute_percentage_error(y_test,pred)
score = max(0, 100*(1-mape))
print('Score :', score)
mse = mean_squared_error(y_test,pred)
print('RMSE :', np.sqrt(mse))
```

## Gradient Boosting regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)
pred = gb.predict(x_test)
mape = mean_absolute_percentage_error(y_test,pred)
score = max(0, 100*(1-(mape)))
print('Score :', score)
print(mean_absolute_error(y_test,pred))
mse = mean_squared_error(y_test,pred)
print('RMSE :', np.sqrt(mse))
```

## XGBT

```python
from xgboost import XGBRegressor
xgbt = XGBRegressor()
xgbt.fit(x_train,y_train)
pred = xgbt.predict(x_test)
mape = mean_absolute_percentage_error(y_test,pred)
score = max(0, 100*(1-mape))
print('Score :', score)
mse = mean_squared_error(y_test,pred)
print('RMSE :', np.sqrt(mse))
```

### Hyperparameter tuning Random forest regressor model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rlf=RandomForestRegressor()
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.10,random_state=478)
params = {'n_estimators':[120,140],
        'min_samples_split':[1,8],
        'min_samples_leaf':[1,5],
        'max_features': range(3,6)
        }
grd = GridSearchCV(rlf,param_grid = params)
grd.fit(x_train,y_train)
print('best_pram', grd.best_params_)

rf=grd.best_estimator_ #reinstantiating with best params

rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)

mape = mean_absolute_percentage_error(y_test,y_pred)
score = max(0, 100*(1-mape))
print('Score :', score)
mse = mean_squared_error(y_test,y_pred)
print('RMSE :', np.sqrt(mse))
```
